```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
params = {
    'axes.labelsize': 10,
    'legend.fontsize': 10,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'text.usetex': True,
    'figure.figsize': [7, 3.375],
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'ytick.right': True,
    'xtick.top': True
   }
plt.rcParams.update(params)
```

```python
# CR qubits
# ring of qubits: 8.25 MHz
# square: 3.6
# hexagon: 4 
# heavy hexagon: 8
fs = 8


r = [[8.25, 8.4, 5.6, 3.6],
     [6.75, 6.5, 0, 0],
      [0, 5, 7, 0],
#      [2,2,2,2]
    ]

labels = ["qubit CR", "qutrit CR", "qubit CZ"]

ntypes = len(r)

c = [2, 2.4, 3, 4]

scaling = 2.0

fig, ax = plt.subplots(figsize=(scaling*3.37, scaling*3))
# ax.grid()

for i in range(ntypes):
    ax.bar([k*(ntypes+1) + i for k in range(4)], r[i], label=labels[i])



ax.set_xticks([k*(ntypes+1)+len(r)/2 - 0.5 for k in range(4)])
ax.set_xticklabels(['Linear', "Heavy-Hex", "Hexagon", "Square"], fontsize=scaling*fs)
ax.set_ylim(0, 40)
ax.set_ylabel("Tolerance (MHz)", fontsize=scaling*fs)

ax.axhline(15, ls='-.',lw=2, color='Gray', label = r"J.B. Hertzberg $\it{et~al.}$")
# ax.axhline(50, ls='--', color='Gray', label = r"JM. Kreikebaum $\it{et~al.}$")

ax.legend(fontsize=scaling*fs,loc=2)

ax.text(0.5, 0.5, 'PLACEHOLDER', transform=ax.transAxes,
        fontsize=scaling*30, color='red', alpha=0.5,
        ha='center', va='center', rotation='30')

fig.savefig("lattice_yield.pdf")
```
